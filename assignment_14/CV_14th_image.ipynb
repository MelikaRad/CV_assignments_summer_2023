### rotating 9 degrees and transforming 10 units to right 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

rotation_amount_degree = 9
#  convert rotation amount to radian
theta = rotation_amount_degree * np.pi / 180.0


image = cv2.imread('iloveyouticket.jpg')
height, width, _ = image.shape

# rotating 20 degrees and transforming 10 units to right 
T = np.float32([[np.cos(theta), -np.sin(theta), 10], [np.sin(theta), np.cos(theta),0]])
img_translation = cv2.warpAffine(image, T, (width+10, height+50))
plt.imshow(img_translation[...,::-1])
```
