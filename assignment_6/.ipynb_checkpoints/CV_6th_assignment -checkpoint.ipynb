```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 

# Making The Blank Image
img = np.zeros((512,512,3))
drawing = False
b,g,r = 0, 0, 0
ix = 0
iy = 0

# Adding Function Attached To Mouse Callback
def draw(event,x,y,flags,params):
    global ix,iy,drawing,b,g,r
    # Left Mouse Button Down Pressed
    if event==cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix = x
        iy = y
    if event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            #For Drawing Line
            cv2.line(img,pt1=(ix,iy),pt2=(x,y),color=(b,g,r),thickness=3)
            ix = x
            iy = y
    if event==cv2.EVENT_LBUTTONUP:
        drawing = False


# Making Window For The Image
cv2.namedWindow('image')
  
def change_red(val):
    global r 
    r = val
    
def change_green(val):
    global g
    g = val
    
def change_blue(val):
    global b 
    b = val
    
    
# Making Window For The Image
cv2.namedWindow('image')

# Adding Mouse CallBack Event
cv2.setMouseCallback('image',draw)

    
    
# create trackbars for color change
cv2.createTrackbar('R','image',0,255,change_red)
cv2.createTrackbar('G','image',0,255,change_green)
cv2.createTrackbar('B','image',0,255,change_blue)


while True:
    cv2.imshow('image',img)
    if cv2.waitKey(1) & 0xFF ==27:
        break
    # get current positions of four trackbars
    color=[b,g,r]
    
cv2.destroyAllWindows()
print('last b is ' + str(b))
print('last g is ' + str(g))
print('last r is ' + str(r))
plt.imshow(img[...,::-1])
```

```python

import matplotlib.pyplot as plt
import numpy as np
import cv2 

# Making The Blank Image
img = np.zeros((512,512,3))
img.fill(0)

drawing = False
b,g,r = 0, 0, 0
colorr = (0,0,0)
ix = 0
iy = 0

# Adding Function Attached To Mouse Callback
def draw(event,x,y,flags,params):
    global ix,iy,drawing,color,b,g,r
    # Left Mouse Button Down Pressed
    if event==cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix = x
        iy = y
    if event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            #For Drawing Line
            cv2.line(img, (ix,iy),(x,y),colorr,3)
            ix = x
            iy = y
    if event==cv2.EVENT_LBUTTONUP:
        drawing = False


# Making Window For The Image
cv2.namedWindow('image')

# Adding Mouse CallBack Event
cv2.setMouseCallback('image',draw)

    
def nothing(x):
    pass
    
# create trackbars for color change
cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)


while True:
    cv2.imshow('image',img)
    if cv2.waitKey(1) & 0xFF ==27:
        break
    # get current positions of four trackbars
    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G','image')
    b = cv2.getTrackbarPos('B','image')
    
    
    colorr=(b,g,r)
    
cv2.destroyAllWindows()
print('last b is ' + str(b))
print('last g is ' + str(g))
print('last r is ' + str(r))
print(colorr)
plt.imshow(img[...,::-1])
```
