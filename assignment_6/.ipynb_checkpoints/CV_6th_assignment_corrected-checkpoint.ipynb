this simple paint application gives user the following tools.  

the ability to draw by pen, if the shape trackbar is set to 0.  
shape == 0  

the ability to draw a rectangle, if the shape trackbar is set to 1.     
shape == 1  

the ability to draw an ellipse, if the shape trachbar is set to 2.  
shape == 2

the ability to decide whether or not the shape is filled, using filled trackbar.   

the ability to change color of the shapes.  
R,G,B 

the ability to change thickness of the shapes, using thickness trackbar.    
thickness



```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 

# Making white Blank Image
width, height = 800, 600
img = np.ones((height, width, 3), np.uint8) * 255

# img = np.zeros((512,512,3)) # i have no idea why it doesn't work on a black image !! the colors will be messed :(((


start_point = (0,0)
points = []      
drawing = False
ix = 0
iy = 0
move=False

# Adding Function Attached To Mouse Callback
def draw(event,x,y,flags,params):
    global ix,iy,drawing,start_point, points, move
    
    if cv2.getTrackbarPos('shape','image')==0 : # pen
        
        # Left Mouse Button Down Pressed
        if event==cv2.EVENT_LBUTTONDOWN:
            drawing = True
            ix = x
            iy = y
            
            # drawing a dot even if the mouse didn't move 
            cv2.circle(img, (x, y), int(cv2.getTrackbarPos('thickness','image')/1.7) ,   # got the 1.7 with trial and error :))
                      color=(cv2.getTrackbarPos('B','image'),
                                cv2.getTrackbarPos('G','image'),
                                cv2.getTrackbarPos('R','image')), 
                      thickness=-2)
            
            
        if event==cv2.EVENT_MOUSEMOVE: # line
            
            if drawing==True:
                #For Drawing Line
                cv2.line(img,pt1=(ix,iy),pt2=(x,y),
                         color=(cv2.getTrackbarPos('B','image'),
                                cv2.getTrackbarPos('G','image'),
                                cv2.getTrackbarPos('R','image')),
                         thickness= cv2.getTrackbarPos('thickness','image'))
                ix = x
                iy = y
                
        if event==cv2.EVENT_LBUTTONUP:   
            drawing = False 
            
            

            
    elif cv2.getTrackbarPos('shape','image')==1 : # rectangle 
    
        if event == cv2.EVENT_LBUTTONDOWN:
            start_point = (x, y)
            drawing = True 
        elif event == cv2.EVENT_MOUSEMOVE:
            if drawing:
                temp_image = img.copy()
                cv2.rectangle(temp_image, start_point, (x, y), (cv2.getTrackbarPos('B','image'),
                                                                cv2.getTrackbarPos('G','image'),
                                                                cv2.getTrackbarPos('R','image')), 
                              thickness= pow(-1,cv2.getTrackbarPos('filled','image'))*cv2.getTrackbarPos('thickness','image'))
                cv2.imshow("image", temp_image)
                
        # check to see if the left mouse button was released
        elif event == cv2.EVENT_LBUTTONUP:
            # record the ending (x, y) coordinates and indicate that
            # the cropping operation is finished
            points.append([start_point , (x, y)])

            # draw a rectangle around the region of interest
            cv2.rectangle(img, start_point, (x, y), (cv2.getTrackbarPos('B','image'),
                                                     cv2.getTrackbarPos('G','image'),
                                                     cv2.getTrackbarPos('R','image')), 
                          thickness= pow(-1,cv2.getTrackbarPos('filled','image'))*cv2.getTrackbarPos('thickness','image'))
            cv2.imshow("image", img)
            drawing = False
            
            
            
            
            
    elif cv2.getTrackbarPos('shape','image')==2 : # ellipse 
    
        if event == cv2.EVENT_LBUTTONDOWN:
            start_point = (x, y)
            drawing = True 
        elif event == cv2.EVENT_MOUSEMOVE:
            if drawing:
                temp_image = img.copy()
                cv2.ellipse(temp_image,
                            (abs((x+start_point[0])//2), abs((y+start_point[1])//2)) ,
                            (abs((x-start_point[0])//2), abs((y-start_point[1])//2)), 
                            0,0,360,
                            (cv2.getTrackbarPos('B','image'),
                             cv2.getTrackbarPos('G','image'),
                             cv2.getTrackbarPos('R','image')), 
                            thickness= pow(-1,cv2.getTrackbarPos('filled','image'))*cv2.getTrackbarPos('thickness','image'))
                cv2.imshow("image", temp_image)
        # check to see if the left mouse button was released
        elif event == cv2.EVENT_LBUTTONUP:
            # record the ending (x, y) coordinates and indicate that
            # the cropping operation is finished
            points.append([start_point , (x, y)])

            # draw an ellipse around the region of interest
            cv2.ellipse(img,
                        (abs((x+start_point[0])//2), abs((y+start_point[1])//2)),
                        (abs((x-start_point[0])//2), abs((y-start_point[1])//2)), 
                        0,0,360,
                        (cv2.getTrackbarPos('B','image'),
                         cv2.getTrackbarPos('G','image'),
                         cv2.getTrackbarPos('R','image')), 
                        thickness= pow(-1,cv2.getTrackbarPos('filled','image'))*cv2.getTrackbarPos('thickness','image'))
            cv2.imshow("image", img)
            drawing = False
            
        



# Making Window For The Image
cv2.namedWindow('image')

def nothing(x):
    pass
    
# create trackbars for color change, thickness, and type of shape
cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)
cv2.createTrackbar('thickness','image',1,20,nothing)
cv2.createTrackbar('shape','image',0,2,nothing)
cv2.createTrackbar('filled','image',0,1,nothing)


# Adding Mouse CallBack Event
cv2.setMouseCallback('image', draw )

    
def nothing(x):
    pass


while True:
    cv2.imshow('image',img)
    if cv2.waitKey(2) & 0xFF ==27:
        break
    
    
cv2.destroyAllWindows()

plt.imshow(img[...,::-1])
```
