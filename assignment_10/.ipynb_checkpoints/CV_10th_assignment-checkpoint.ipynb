```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img =  cv2.imread('adad.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

noise_removed_img = cv2.GaussianBlur(gray, (9,9), 0)

binary_img = cv2.adaptiveThreshold(noise_removed_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,5,2)
plt.imshow(binary_img, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
# drawing contours 
# (not wanted for the task btw. )

img2 = img.copy()

for c in contours:
    cv2.drawContours(img2, [c], -1, (255,0,0), 3)
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
# drawing centers

img3 = img.copy()

centers = []
for c in contours:
    M = cv2.moments(c)
    cx = int(M['m10'] / M['m00'])
    cy = int(M['m01'] / M['m00'])
    cv2.circle(img3, (cx, cy), 5, (255,0,0), -1)
    centers.append((cx, cy)) # for later use

centers
```

```python
plt.imshow(img3[:,:,::-1])
```

```python
# sorting centers based on x axis, to find the rightmost center

sorted_centers = sorted(centers, key=lambda p: p[0], reverse=True)  # sort based on first element
print(sorted_centers)
```

```python
# sorting centers based on y axis, to find the topmost center

sorted_centers = sorted(centers, key=lambda p: p[1], reverse=False)  # sort based on second element
print(sorted_centers)
```

```python
# funtions for sorting contours based on their x axis (reversed) or y axis

def get_contour_x(contour):
    M = cv2.moments(contour)
    cx = int(M["m10"] / M["m00"])
    return cx


def get_contour_y(contour):
    M = cv2.moments(contour)
    cy = int(M["m01"] / M["m00"])
    return cy

def sort_contours_by_x(contours):
    return sorted(contours, key=lambda c: get_contour_x(c), reverse=True)

def sort_contours_by_y(contours):
    return sorted(contours, key=lambda c: get_contour_y(c))
```

```python
# sorted contours, both by x(reversed) and y

sorted_contours_by_x = sort_contours_by_x(contours)
sorted_contours_by_y = sort_contours_by_y(contours)
```

#### Rightmost digit :

```python
# drawing rectangle around the rightmost digit 

img5 = img.copy()
boundRect = cv2.boundingRect(sorted_contours_by_x[0])
cv2.rectangle(img5, (int(boundRect[0]), int(boundRect[1])), \
             (int(boundRect[0]+boundRect[2]), int(boundRect[1]+boundRect[3])), (0,255,255), 2)
plt.imshow(img5[:,:,::-1])
```

#### Topmost digit :

```python
# drawing rectangle around the topmost digit 

img5 = img.copy()
boundRect = cv2.boundingRect(sorted_contours_by_y[0])
cv2.rectangle(img5, (int(boundRect[0]), int(boundRect[1])), \
             (int(boundRect[0]+boundRect[2]), int(boundRect[1]+boundRect[3])), (0,255,255), 2)
plt.imshow(img5[:,:,::-1])
```
