### by histogram equalization

#### global histogram equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('image.png', 0)
result = cv2.equalizeHist(image)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("After global histogram equalization");
```

#### local histogram equalization

```python
clahe = cv2.createCLAHE(clipLimit=50, tileGridSize=(100,100))
result = clahe.apply(image)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("After adaptive histogram equalization");
```

```python
clahe = cv2.createCLAHE(clipLimit=100, tileGridSize=(50,50))
result = clahe.apply(image)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("After adaptive histogram equalization");
```

### by pixel multiplication 

```python
image = cv2.imread('image.png', 0)
result = np.clip(image*55, 0, 255)

plt.figure(figsize=[12,4])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(result, cmap='gray');plt.title("multiplied image");
```
