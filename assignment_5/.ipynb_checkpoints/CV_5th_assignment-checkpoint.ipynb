```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# رسم مثلث هایی با دو رنگ و دو چرخش متفاوت

colors = [(255, 20, 255), (150, 255, 20)]
color_index = 0
rotate = 0

def draw_triangle(event, x, y, flags, param):
    
    global color_index, rotate
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if rotate == 1:
            vertices = np.array([[x, y-30], [x+30, y+30], [x-30, y+30]])
        else:
            vertices = np.array([[x, y+30], [x-30, y-30], [x+30, y-30]])

        cv2.fillPoly(img, [vertices], colors[color_index]) 
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        color_index = (color_index + 1) % 2
        rotate = (rotate+1) % 2


img = np.zeros((512, 512, 3), np.uint8)

cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_triangle)

while True:
    cv2.imshow('image', img)
    key = cv2.waitKey(1) & 0xFF
    if key == 27:
        break

cv2.destroyAllWindows()
plt.imshow(img[...,::-1])
```
