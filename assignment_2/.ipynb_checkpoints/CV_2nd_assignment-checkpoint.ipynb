```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('erik_demaine.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(RGB_img)
```

```python
img.shape
```

```python
RGB_img.shape
```

```python
circle = np.zeros(RGB_img.shape, np.uint8)
cv2.circle(circle, (400, 140), 100, (255,255,255), -1) 

masked = cv2.bitwise_and(RGB_img, circle)
```

```python
plt.imshow(masked)
```

```python
# هدف از این قسمت صرفا این بود که اندازه و مکان دایره دور صورت مشخص بشه 
```

making 4-channel png image and showing it by plt:

```python
four_channel_img = np.zeros((350,579,4), np.uint8)
```

```python
alpha = np.zeros((350,579), np.uint8)
cv2.circle(alpha, (400, 140), 100, 255, -1) 
alpha[alpha==0] = 100


four_channel_img[:,:,:3] = RGB_img
four_channel_img[:,:, 3] = alpha
```

```python
plt.imshow(four_channel_img)
```

```python
# again... creating the image and saving it  

alpha = np.zeros((350,579), np.uint8)
cv2.circle(alpha, (400, 140), 100, 255, -1) 
alpha[alpha==0]=100


four_channel_img[:,:,:3] = img # (no channel conversion at this part)

four_channel_img[:,:, 3] = alpha

cv2.imwrite('ConcreteErikFace_FadedBackground.png', four_channel_img)
```
