```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((400, 400, 3), dtype=np.uint8)
image[...,0] = 250
image[...,1] = 150
image[...,2] = 100


cv2.rectangle(image, (100, 200), (300, 400), (0, 150, 200), -1) # بدنه 

pts = np.array([[100, 200], [200, 100], [300, 200]], np.int32)
cv2.fillPoly(image, [pts], (0, 100, 160)) # شیروانی 

cv2.line(image, (75,225), (200,100), (0,70,90), 8) # سقف شیروانی
cv2.line(image, (325,225), (200,100), (0,70,90), 8) # سقف شیروانی 


cv2.rectangle(image, (170, 320), (230, 400), (0, 100, 160), -1) # در

cv2.circle(image, (220,360), 5, (0,70,90), -1)

cv2.rectangle(image, (120, 240), (170, 290), (250, 170, 140), -1) # پنجره چپ
cv2.rectangle(image, (230, 240), (280, 290), (250, 170, 140), -1) # پنجره راست


cv2.imshow("House", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB ))
```
