## Gamma Correction :

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("Weeki_Wachee_spring_10079u.jpg")
```

```python
# for gamma = 2

gamma = 2
lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
gamma_eq_2 = cv2.LUT(img_original, lookUpTable)
```

```python
# for gamma = 1

gamma = 1
lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
gamma_eq_1 = cv2.LUT(img_original, lookUpTable)
```

```python
# for gamma = 0.5

gamma = 0.5
lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
gamma_eq_half = cv2.LUT(img_original, lookUpTable)
```

```python
# for gamma = 1/3

gamma = (1/3)
lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
gamma_eq_1over3 = cv2.LUT(img_original, lookUpTable)
```

```python
# for gamma = 0.25

gamma = 0.25
lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
gamma_eq_1over4 = cv2.LUT(img_original, lookUpTable)
```

```python
plt.figure(figsize=[20,15])
plt.subplot(231);plt.imshow(img_original[...,::-1]);plt.title("Original");
plt.subplot(232);plt.imshow(gamma_eq_2[...,::-1]);plt.title("gamma = 2");
plt.subplot(233);plt.imshow(gamma_eq_1[...,::-1]);plt.title("gamma = 1 , same as original");
plt.subplot(234);plt.imshow(gamma_eq_half[...,::-1]);plt.title("gamma = 1/2");
plt.subplot(235);plt.imshow(gamma_eq_1over3[...,::-1]);plt.title("gamma = 1/3");
plt.subplot(236);plt.imshow(gamma_eq_1over4[...,::-1]);plt.title("gamma = 1/4");
```

# Linear Adjustment:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("Weeki_Wachee_spring_10079u.jpg")

new_image = np.zeros(image.shape, image.dtype)

alpha = np.arange(1,10,2)
beta = np.arange(0,100,30)

fig, ax = plt.subplots(5,4, figsize=(20,20))


for i in range(5):
    for j in range(4):
        new_image = cv2.convertScaleAbs(image, alpha=alpha[i], beta=beta[j])
        ax[i,j].imshow(new_image)
        ax[i,j].set_title('alpha='+str(alpha[i])+ ', beta= '+str(beta[j]))
```
