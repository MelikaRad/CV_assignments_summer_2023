```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### results after making enough trials and errors :

with :   
GaussianBlur kernel size = 13   
adaptiveThreshold kernel size = 7 for mean, 11 for gaussian  

```python
img =  cv2.imread('adad.jpg',0)
noise_removed_img = cv2.GaussianBlur(img, (13,13), 0)
```

```python

# global binary threshol and adaptive binary threshold, on the original image
_,output1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
output2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,7,2)
output3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,7,2)


# global binary threshol and adaptive binary threshold, on the noise-removed image
_,output4 = cv2.threshold(noise_removed_img,127,255,cv2.THRESH_BINARY)
output5 = cv2.adaptiveThreshold(noise_removed_img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,7,2)  # *** 
output6 = cv2.adaptiveThreshold(noise_removed_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)  # *** 


plt.figure(figsize=[15,15])
plt.subplot(331);plt.imshow(img, cmap='gray');plt.title("Original");
plt.subplot(332);plt.imshow(noise_removed_img, cmap='gray');plt.title("Noise removed");

plt.subplot(334);plt.imshow(output1, cmap='gray');plt.title("Global Thresholding (v = 127)");
plt.subplot(335);plt.imshow(output2, cmap='gray');plt.title("Adaptive Mean Thresholding");
plt.subplot(336);plt.imshow(output3, cmap='gray');plt.title("Adaptive Gaussian Thresholding");

plt.subplot(337);plt.imshow(output4, cmap='gray');plt.title("Global Thresholding + noise removal");
plt.subplot(338);plt.imshow(output5, cmap='gray');plt.title("Adaptive Mean + noise removal");
plt.subplot(339);plt.imshow(output6, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
```
